# Exploring the MSRC Dataset

### 1. Environment Setup

```python
import cv2
import os
import sys
import numpy as np
import scipy.io as sio

from dotenv import load_dotenv
load_dotenv()
```

#### 2. Load the Data

```python
BASE_PATH = os.getenv("BASE_PATH")
print(f"BASE_PATH: {BASE_PATH}")

MSRC_FOLDER = os.path.join(BASE_PATH, 'MSRC_ObjCategImageDatabase_v2')
MSRC_SUBFOLDER = 'Images'

MSRC_directory = os.path.join(MSRC_FOLDER, MSRC_SUBFOLDER)
print(f"MSRC_directory: {MSRC_directory}")
```

#### 3. Helper Function

```python
def extract_image_labels(path: str):
    image_id = os.path.splitext(os.path.basename(path))[0]
    class_part = image_id.split("_")[0]
    try:
        class_id = int(class_part)
    except ValueError as exc:
        raise ValueError(f"Cannot parse class id from {image_id}") from exc
    return image_id, class_id
```

#### 4. Get some basic counts & save to CSV

```python
ground_truth = []

img_files = sorted(
    f for f in os.listdir(MSRC_directory)
    if f.lower().endswith(".bmp")
)
if not img_files:
    raise RuntimeError(f"No .bmp image files found under {MSRC_directory}")

for filename in img_files:
    ground_truth.append(extract_image_labels(filename))

print(ground_truth[20:80])
```

```python
import csv

output_csv_path = os.path.join(BASE_PATH, "ground_truth_labels.csv")

with open(output_csv_path, mode="w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["image_id", "class_id"])
    for image_id, class_id in ground_truth:
        writer.writerow([image_id, class_id])

print(f"Saved ground truth data to {output_csv_path}")
```
